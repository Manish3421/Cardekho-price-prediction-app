```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = r"C:\Users\manis\OneDrive\Desktop\final_car.xlsx"
df = pd.read_excel(file_path)

# Check column names to confirm correct column names
print("Columns in the dataset:")
print(df.columns)

# Basic Overview of the Data
print(f"Shape of the data: {df.shape}")
print(df.head())  # Checking the first few rows
print(df.info())  # Data types and non-null counts
print(df.describe())  # Summary statistics

# Missing Values
missing_values = df.isnull().sum()
print("\nMissing values:")
print(missing_values[missing_values > 0])  # Display columns with missing values

# Distribution of Numerical Features
numerical_cols = ['price', 'Mileage', 'Seats']  # Corrected to use 'Mileage'
df[numerical_cols].hist(figsize=(10, 8))
plt.suptitle("Distribution of Numerical Features")
plt.show()

# Categorical Features Analysis
categorical_cols = ['transmission', 'owner', 'ft', 'oem', 'model']  # Corrected 'fuelType' to 'ft'
for col in categorical_cols:
    plt.figure(figsize=(8, 6))
    sns.countplot(y=df[col])
    plt.title(f'Distribution of {col}')
    plt.show()

# Clean the Mileage column by removing non-numeric characters and converting to float
df['Mileage'] = df['Mileage'].replace(r'[^\d.]', '', regex=True).astype(float)

# Now proceed with the correlation matrix
correlation_matrix = df[numerical_cols].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()

```
